# 1. Matrix

In this course and a large variety of practical settings of data analysis, data are stored in matrices. Define $A_{n\times m}=(A_{i,j})_{n\times m}$ as a matrix of $n$ rows and $m$ columns, where $n$ and $m$ are positive integers. For example, below we have a 2-by-3 matrix $A$:
$$A=\left(\begin{matrix}
A_{11}&A_{12}&A_{13}\\
A_{21}&A_{22}&A_{23}
\end{matrix}\right)=\left(\begin{matrix}
2&0&-1\\
3&1&2
\end{matrix}\right)$$
In practice, the dimension of a matrix can be huge as we are in a big-data array. 

We say two matrices $A=B$ if each entry of $A$ is the same as the corresponding entry of $B$. That is
$$A_{ij}=B_{ij}\mbox{ for all }1\le i\le n,1\le j\le m$$

### Matrix Operations

We describe some commonly used matrix operations. The first one is transpose. We denote the transpose of matrix $A$ as $A^T$. If $B=A^T$, then $B_{ij}=A_{ji}$ for any $i$ and any $j$. Therefore, if $A$ is $n$ by $m$, then $A^T$ is $m$ by $n$. For example, if 
$$A=\left(\begin{matrix}
2&0&-1\\
3&1&2
\end{matrix}\right),\mbox{ then }A^T=\left(\begin{matrix}
2&3\\
0&1\\
-1&2
\end{matrix}\right)$$

The second operation for matrix is summation. Assume $A$ and $B$ are two matrices of the same dimensions ($n\times m$), then the sum of $A$ and $B$ is an $n$ by $m$ matrix satisfying
$$A+B=\left(\begin{matrix}
A_{11}&A_{12}&...&A_{1m}\\
A_{21}&A_{22}&...&A_{2m}\\
...&...&...&...\\
A_{n1}&A_{n2}&...&A_{nm}
\end{matrix}\right)+\left(\begin{matrix}
B_{11}&B_{12}&...&B_{1m}\\
B_{21}&B_{22}&...&B_{2m}\\
...&...&...&...\\
B_{n1}&B_{n2}&...&B_{nm}
\end{matrix}\right)=\left(\begin{matrix}
A_{11}+B_{11}&A_{12}+B_{12}&...&A_{1m}+B_{1m}\\
A_{21}+B_{21}&A_{12}+B_{22}&...&A_{1m}+B_{2m}\\
...&...&...&...\\
A_{n1}+B_{n1}&A_{n2}+B_{n2}&...&A_{nm}+B_{nm}
\end{matrix}\right)$$
For example,
$$\left(\begin{matrix}
2&0&-1\\
3&1&2
\end{matrix}\right)+\left(\begin{matrix}
-1&3&1\\
2&-2&4
\end{matrix}\right)=\left(\begin{matrix}
1&3&0\\
5&-1&6
\end{matrix}\right)$$
The matrix substraction operation can be defined as:
$$A-B=A+(-B)$$
where the dimensions of $A$ and $B$ are the same.

Similarly, we can define the scalar multiplication of a matrix, which is a matrix of the same dimension: if $$A=\left(\begin{matrix}
A_{11}&A_{12}&...&A_{1m}\\
A_{21}&A_{22}&...&A_{2m}\\
...&...&...&...\\
A_{n1}&A_{n2}&...&A_{nm}
\end{matrix}\right)$$ 
then
$$kA=\left(\begin{matrix}
kA_{11}&kA_{12}&...&kA_{1m}\\
kA_{21}&kA_{22}&...&kA_{2m}\\
...&...&...&...\\
kA_{n1}&kA_{n2}&...&kA_{nm}
\end{matrix}\right)$$
For example, 
$$A=2\times\left(\begin{matrix}
2&0&-1\\
3&1&2
\end{matrix}\right)=\left(\begin{matrix}
4&0&-2\\
6&2&4
\end{matrix}\right)$$



Next, we consider the multiplication of matrices. Consider two matrices $A$ and $B$, where $A$ is $p$ by $q$ and $B$ is $q$ by $r$. Then, $M=AB$ is a $p$ by $r$ matrix defined as:
$$M_{ij}=\sum_{k=1}^qA_{ik}B_{kj}\mbox{ for all }i\mbox{ and }j$$
For example, if $A=\left(\begin{matrix}
2&0&-1\\
3&1&2
\end{matrix}\right)$ and $B=\left(\begin{matrix}
1&-4\\
2&-1\\
3&-2
\end{matrix}\right)$, then
$$AB=\left(\begin{matrix}
2\times1+0\times2+(-1)\times3&2\times(-4)+0\times(-1)+(-1)\times(-2)\\
3\times1+1\times2+2\times3&3\times(-4)+1\times(-1)+2\times(-2)
\end{matrix}\right)=\left(\begin{matrix}
-1&-6\\
11&-17
\end{matrix}\right)$$

One should note that, in general $AB\ne BA$. In fact, their dimensions may not even match. For example, still assume $A=\left(\begin{matrix}
2&0&-1\\
3&1&2
\end{matrix}\right)$ and $B=\left(\begin{matrix}
1&-4\\
2&-1\\
3&-2
\end{matrix}\right)$. Then, $AB=\left(\begin{matrix}
-1&-6\\
11&-17
\end{matrix}\right)$ and $BA=\left(\begin{matrix}
-10&-4&-9\\
1&-1&-4\\
0&-2&-7
\end{matrix}\right)$.


Matrices enable us to streamline the representations of functions, ineuqlities and equalities. For example:

- $f(x)=\sum_{i=1}^nc_ix_i$. Then, we can write $c=(c_1,c_2,...,c_n)^T$ and $x=(x_1,x_2,...,x_n)^T$. We have $f(x)=c^Tx$. For instance, if $f(x)=2x_1+3x_2$, then $f(x)=(2,3)\left(\begin{matrix}
x_1\\
x_2
\end{matrix}\right)$
- $\sum_{j=1}^nA_{ij}x_j\le B_i$ for $i=1,2,...,m$. We can write $A=(A_{ij})_{m\times n}$, $x=(x_1,x_2,...,x_n)^T$, and $B=(B_1,B_2,...,B_m)$. We have $Ax\le B$. For instance, $2x_1+x_2\le 3$ and $3x_1-2x_2\le 1$ can be written in matrix form as $\left(\begin{matrix}
2&1\\
3&-2
\end{matrix}
\right)\left(\begin{matrix}
x_1\\
x_2
\end{matrix}\right)\le\left(\begin{matrix}
3\\
1
\end{matrix}\right)$.
- $\sum_{j=1}^na_{ij}x_j\le b_i$ for $i=1,2,...,m$. We can write $a=(A_{ij})_{m\times n}$, $x=(x_1,x_2,...,x_n)^T$, and $b=(b_1,b_2,...,b_m)$. We have $ax= B$. For instance, $x_1-2x_2= 1$ and $3x_1+2x_2= 4$ can be written in matrix form as $\left(\begin{matrix}
1&-2\\
3&2
\end{matrix}
\right)\left(\begin{matrix}
x_1\\
x_2
\end{matrix}\right)=\left(\begin{matrix}
1\\
4
\end{matrix}\right)$.

Throughout this course, we will work extensively with matrices to represent and analyze data. In Python, matrices are stored in the format of arrays, using the NumPy and Pandas libraries.

### Fundamental Structures in Computer Programming

We introduce two important structures in computer programming. The first is the branching strucuture, which prescirbes what computations to execute under different conditions.

<img src="branching.png" width=500>


Let's see an application of the branching structure in Python.

````python
def solve_quadratic(a, b, c):
    """
    Compute the real solutions of a quadratic equation ax^2 + bx + c = 0.

    :param a: Coefficient of x^2
    :param b: Coefficient of x
    :param c: Constant term
    :return: A tuple containing the two real solutions, if they exist, or an appropriate message.
    """
    # Case when it's a linear equation (a = 0)
    if a == 0:
        if b == 0:
            return "No equation to solve." if c != 0 else "Infinite solutions."
        else:
            return -c / b, "Linear equation."

    # Calculate the discriminant
    discriminant = b**2 - 4*a*c

    # Check for real solutions
    if discriminant < 0:
        return "No real solutions."
    elif discriminant == 0:
        x = -b / (2 * a)
        return x, x
    else:
        sqrt_discriminant = math.sqrt(discriminant)
        x1 = (-b + sqrt_discriminant) / (2 * a)
        x2 = (-b - sqrt_discriminant) / (2 * a)
        return x1, x2

# Unit tests for the solve_quadratic function
def test_solve_quadratic():
    assert solve_quadratic(1, -3, 2) == (2.0, 1.0), "Test failed for input (1, -3, 2)"
    assert solve_quadratic(1, 4, 4) == (-2.0, -2.0), "Test failed for input (1, 4, 4)"
    assert solve_quadratic(1, 0, -4) == (2.0, -2.0), "Test failed for input (1, 0, -4)"
    assert solve_quadratic(0, 2, -8) == (4.0, "Linear equation."), "Test failed for linear equation (0, 2, -8)"
    assert solve_quadratic(1, 0, 4) == "No real solutions.", "Test failed for no real solutions (1, 0, 4)"
    assert solve_quadratic(0, 0, 5) == "No equation to solve.", "Test failed for no equation (0, 0, 5)"
    assert solve_quadratic(0, 0, 0) == "Infinite solutions.", "Test failed for infinite solutions (0, 0, 0)"
    
    print("All tests passed!")

# Run the unit tests
test_solve_quadratic()
print(solve_quadratic(1,4,3))
````

The `math` library in Python provides the basic mathematical functions (see [here](https://docs.python.org/3/library/math.html)). Here, we are using the function to take the square root `math.sqrt()`. 

The second structure is the for loop strucuture, which prescirbes the repeated computations with respect to an index in the for loop.

<img src="for_loop.png" width=500>


Let's see an application of the for loop structure in Python.

````python
def sum_of_sequence(n, a1, a2):
    """
    Compute the sum of the first n terms of a sequence where each term is the sum of the two preceding terms.

    :param n: The number of terms to sum
    :param a1: The first term of the sequence
    :param a2: The second term of the sequence
    :return: The sum of the first n terms
    """
    if n == 1:
        return a1
    if n == 2:
        return a1 + a2

    sum_seq = a1 + a2
    prev, curr = a1, a2

    for _ in range(3, n + 1):
        next_term = prev + curr
        sum_seq += next_term
        prev, curr = curr, next_term

    return sum_seq

# Unit tests for the sum_of_sequence function
def test_sum_of_sequence():
    assert sum_of_sequence(1, 1, 1) == 1, "Test failed for n=1 with (1, 1)" # Sequence: 1, sum = 1
    assert sum_of_sequence(2, 1, 1) == 2, "Test failed for n=2 with (1, 1)" # Sequence: 1, 1, sum =2
    assert sum_of_sequence(5, 1, 1) == 12, "Test failed for n=5 with (1, 1)"  # Fibonacci sequence: 1, 1, 2, 3, 5, sum = 12
    assert sum_of_sequence(4, 2, 3) == 18, "Test failed for n=4 with (2, 3)"  # Sequence: 2, 3, 5, 8, sum = 18
    assert sum_of_sequence(3, 5, 10) == 30, "Test failed for n=3 with (5, 10)" # Sequence: 5, 10, 15, sum = 30
    assert sum_of_sequence(5, 0, 0) == 0, "Test failed for n=5 with (0, 0)"   # Sequence of zeros, sum = 0
    assert sum_of_sequence(6, -1, 1) == 4, "Test failed for n=6 with (-1, 1)" # Sequence: -1, 1, 0, 1, 1, 2, sum = 4

    print("All tests passed!")

# Run the unit tests
test_sum_of_sequence()

# Example usage
# The sequence starts with 1, 1 and the first five terms are 1, 1, 2, 3, 5. The sum is 12.
print(sum_of_sequence(5, 1, 1))  
````

In Python, `a+ = b` is an augmented assignment operator, which means to assign the value of `a + b` to the variable `a`. Furthermore, `range(a,b)` means the sequence of integers `a, a+1, a+2, ... , b-1` (assuming that `a` and `b` are integers). In the for loop, `_` is a place holder to keep the loop going. 

The final structure is another loop structure called the while loop strucuture, which prescirbes the repeated computations in a loop when the prespecified condition is met.

<img src="while_loop.png" width=500>


Let's see an application of the while loop structure in Python.

````python
def prime_factorization(n):
    """
    Compute the prime factorization of a positive integer.

    :param n: A positive integer
    :return: A list of prime factors
    """
    if n < 1:
        raise ValueError("Input must be a positive integer")
    
    factors = []
    # Factor out the 2s
    while n % 2 == 0:
        factors.append(2)
        n //= 2

    # Factor out other primes
    for i in range(3, int(n**0.5) + 1, 2):
        while n % i == 0:
            factors.append(i)
            n //= i

    # If n is a prime number greater than 2
    if n > 2:
        factors.append(n)

    return factors

# Unit test function for prime factorization.

def test_prime_factorization():
    assert prime_factorization(2) == [2], "Test failed for n=2"
    assert prime_factorization(60) == [2, 2, 3, 5], "Test failed for n=60"
    assert prime_factorization(13) == [13], "Test failed for n=13"  # 13 is a prime number
    assert prime_factorization(100) == [2, 2, 5, 5], "Test failed for n=100"
    try:
        prime_factorization(-10)
        assert False, "Test failed to raise exception for negative input"
    except ValueError:
        pass  # Test passed

    print("All tests passed!")

# Run the unit test
test_prime_factorization()

print(prime_factorization(1001))
````

In Python, `//` means to floor divide, which will return the floor of the quotient, whereas `%` means to obtain the residual. The operator `.append(x)` means to add the element `x` into the list.

# 3. Data Analysis with NumPy

NumPy is a python library for working with arrays. It also has functions convenient for linear algebra, fourier transformation, and matrics operations. In Python, we actually have the data type for arrays (called lists), but they are slow to process. NumPy aims to provide an array object that is up to 50x faster than the traditional Python lists. The array object in NumPy is called ndarray. It provides a lot of supporting functions that facilitates working with ndarray. This section only introduces the very basics. You could refer to numerous online NumPy tutorials to learn more. 

To begin with, we first create 1- and 2-dimensional arrays. The method in NumPy `array()` creates an ndarray object, whereas the method `shape` returns the  lengths of the corresponding array dimensions.

````python
## Create a vector:
vec = np.array([1, 2, 3, 4, 5])

print(vec)
print('The shape of the vector is',vec.shape)

## Create a matrix:

mat = np.array([[1,2,3],[4,5,6]])
print('\n')
print(mat)
print('The shape of the matrix is', mat.shape)
````

Next, we try to access different entries of an array.

````python
# For a vector, the index starts with 0. Index -1 means the last element.

print(vec[0],vec[2],vec[-1])

# For a matrix, the index specifies both the row and the column.

print(mat[0,2])
print(mat[1,:])
print(mat[:,-1])
````

Next, we reshape an array using the method `reshape()`.

````python
# Reshape mat into a vector and into a 1 by 6, and a 3 by 2 matrices.
mat1 = mat.reshape(1,6)
mat2 = mat.reshape(3,2)
print(mat1)
print(mat2)
````

Next, we consider joining multiple arrays into a single one. In NumPy, we use the `concatenate()` function to join the arrays. We also should specify along which axis we want to join the arrays. If axis is not explicitly passed, its default value is 0.

````python
# Join arrays:

arr1 = np.array([[1, 2, 3]])

arr2 = np.array([[4, 5, 6]])

arr_join0 = np.concatenate((arr1, arr2), axis=0)

arr_join1 = np.concatenate((arr1, arr2), axis=1)
print('The array joined along axis 0:','\n',arr_join0)
print('Shape of the array joined along axis 0:',arr_join0.shape)
print('\n')
print('The array joined along axis 1:',arr_join1)
print('Shape of the array joined along axis 1:',arr_join1.shape)
````

Next, we discuss how to join DataFrames using indices. We will use the `merge()` function in Pandas. We want to highlight that joining is arguably the most important operation in data analysis since it combines the information from different data sets in an integral fashion.

````python
## Create 2 DataFrames
df_BA = pd.DataFrame({'Name':['Adam','Bob','Charles','Danniel','Edward','Frank'],\
                      'BA_Grade':['B-','A','A-','B+','B','A']})
df_Finance = pd.DataFrame({'Name':['Amy','Bob','Catherine','Danniel','Edward','Fiona'],\
                      'Finance_Grade':['A','B+','B-','A','B','B-']})
df_BA
````

````python
df_Finance
````

````python
#Outer join

df_outer = pd.merge(df_BA,df_Finance,on = ['Name'],how = 'outer')
df_outer
````

````python
# Left join
df_left = pd.merge(df_BA,df_Finance,on = ['Name'],how = 'left')
df_left
````

````python
# Right join
df_right = pd.merge(df_BA,df_Finance,on = ['Name'],how = 'right')
df_right
````

````python
# Inner join
df_inner = pd.merge(df_BA,df_Finance,on = ['Name'],how = 'inner')
df_inner
````

The next operation of Pandas we want to discuss in this document is the aggregate data using the `group_by()` function. To illustrate this application, we create a DataFrame of T-shirt sales data first. The function `MultiIndex.from_arrays()` creates multiple indices from an array.

````python
# Create a dictionary.
data_dict = {'color' : ['black', 'white', 'black', 'white', 'black',\
                      'white', 'black', 'white', 'black', 'white'],\
            'size' : ['S','M','L','M','L','S','S','XL','XL','M'],\
            'price': [99,79,108,89,99,100,105,111,86,94],\
            'quantity': [1210,1340,2410,1458,1124,2217,989,1009,1342,1077]}
array=[['A','B','B','B','C','A','B','A','C','C'],['JP','CN','US','US','US','CN','CN','CA','JP','CA']]

# Create the two indices.
index = pd.MultiIndex.from_arrays(array, names=['class', 'country'])

# Create the DataFrame.
data_df = pd.DataFrame(data_dict,index=index)
data_df
````

````python
## We group the data by size

df_size = data_df.groupby(["size"]).mean(numeric_only=True)

# Calculate the average price and quantity of each size

df_size.add_prefix('average_')
````

We now summarize other aggregation functions that could be used for the `group_by()` function.

- `count`: Count the number of elements in the group that are not NA.
- `mean`: Compute the average value of the group.
- `median`: Compute the median value of the group.
- `std`: Compute the standard deviation of the group.
- `var`: Compute the variance of the group.
- `min`: Compute the minimum value of the group.
- `max`: Compute the maximum value of the group.
- `prod`: Compute the product of all values in the group.
- `first`: Find the first value in the group that is not NA.
- `last`: Find the last value in the group that is not NA.

# 5. Data Visualization

Data visualization is important for explorative data analysis and for communicating your insights from data analysis. For the interest of time, the introduction will be very general and incomplete. To learn more about this topic, you may refer to the online tutorials of the `Matplotlib` package ([Matplotlib Documentation](https://matplotlib.org/)). We will also cover certain data visualization methods as we progress in this course.

Matplotlib is the basis for static plotting in Python. Many other visualization tools are built on top of it, such as seaborn and Pandas DataFrames plot method. Matplotlib is flexible enough for plot anything but may not be very convenient for complex visualizations.

Let's first try the line plot.

````python
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x1,y,'ys-') # solid line with yellow color and square marker
l2 = ax.plot(x2,y,'go--') # dash line with green color and circle marker
ax.legend(labels = ('tv', 'Smartphone'), loc = 4) # legend placed at lower right
ax.set_title("Effect of advertising on sales")
ax.set_xlabel('Cost')
ax.set_ylabel('Sales')
plt.show()
````

Then, the bar plot.

````python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(N) # the locations for the groups. np.range(N) returns an array [0,1,...,N-1]
width = 0.35 # The width of the bar in the bar plot.
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, menMeans, width, color='r')
ax.bar(ind, womenMeans, width,bottom=menMeans, color='b') #bottom specifies where the bottom of the bar is.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(['G1', 'G2', 'G3', 'G4', 'G5'])
ax.set_yticks(np.arange(0, 81, 10))  
#np.arange(0,81,10) returns an array starting at 0, stopping at 81 (excluded), with a gap of 10
ax.legend(labels=['Men', 'Women'])
plt.show()
````

Finally, we try the scatter plot.

````python
BA_grades = [89, 85, 70]
Finance_grades = [70, 90, 85]
name = ['Adam', 'Bob', 'Charles']
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(name, BA_grades, color='r')
ax.scatter(name, Finance_grades, color='b') 
ax.set_xlabel('Grades')
ax.set_ylabel('Names')
ax.set_title('Scatter Plot')
plt.show()
````

You should also note that the line plot above is a more efficient way of data visualization than the bar plot and the scatter plot. Think about way from the perspective of actionability.
